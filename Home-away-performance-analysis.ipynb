# Calculate the affect of playing home vs away             



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("data/2020-2021.csv")
```

```python
df.head(100)
```

```python
# Win Rate Calculations (considering draws)
total_matches = len(df)
home_wins = df[df['FTR'] == "H"]
away_wins = df[df['FTR'] == "A"]
draws = df[df['FTR'] == "D"]

home_win_rate = (len(home_wins) / total_matches) * 100
away_win_rate = (len(away_wins) / total_matches) * 100
draw_rate = (len(draws) / total_matches) * 100

# Visualizations (formatting for percentages)
plt.figure(figsize=(8, 6))
plt.bar(["Home", "Away", "Draw"], [home_win_rate, away_win_rate, draw_rate])
plt.xlabel("Team/Outcome")
plt.ylabel("Win Rate (%)")  # Update label to show percentage
plt.title("Win Rate (Home vs Away) ")
plt.xticks(rotation=0)
plt.show()
```
