```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv("data/2020-2021.csv")
```

```python
mu_home_matches = data[data['HomeTeam'] == 'Man United']
mu_away_matches = data[data['AwayTeam'] == 'Man United']
mu_matches = pd.concat([mu_home_matches, mu_away_matches])
mu_matches.info()
```
